```python
#Load the data

import scipy.io
import numpy as np

data = scipy.io.loadmat('/Users/kanishksaxena/Downloads/train_32x32.mat')

#Understand the dataset
keys = data.keys()

images = data['X']
labels = data['y']
```

```python
#Data Summary
num_samples = images.shape[3]
num_classes = len(np.unique(labels))
image_shape = images.shape

print(f"Number of samples: {num_samples}")
print(f"Number of classes: {num_classes}")
print(f"Image shape: {image_shape}")
```

```python
#Data Visualization
import matplotlib.pyplot as plt

# Plot sample images
plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(images[:, :, :, i])
    plt.title(f"Label: {labels[i][0]}")
plt.show()
```

```python
#Class Distribution

import numpy as np

label_counts = np.bincount(labels.flatten())
plt.bar(np.arange(1, num_classes + 1), label_counts[1:])
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Distribution')
plt.show()
```

```python
#Data Preprocessing
images = images.transpose((3, 2, 0, 1))  # Reshape for convolutional networks
labels = labels % 10  # Adjust for label 10 (digit 0)

# Normalize images to [0, 1]
images = images / 255.0
```

```python
#Saving preprocessed data
np.savez('svhn_data.npz', images=images, labels=labels)
```
